```python
# 01_data_exploration.ipynb

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '../data/market_mix_data.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()

# Basic statistics of the dataset
df.describe()

# Check for missing values
df.isnull().sum()

# Plot the sales data
plt.figure(figsize=(10, 6))
plt.plot(df['week'], df['sales'], marker='o')
plt.title('Weekly Sales')
plt.xlabel('Week')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

# Histogram of marketing spends
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.hist(df['tv_spend'], bins=20, color='blue', alpha=0.7)
plt.title('TV Spend Distribution')

plt.subplot(2, 2, 2)
plt.hist(df['radio_spend'], bins=20, color='green', alpha=0.7)
plt.title('Radio Spend Distribution')

plt.subplot(2, 2, 3)
plt.hist(df['online_spend'], bins=20, color='red', alpha=0.7)
plt.title('Online Spend Distribution')

plt.subplot(2, 2, 4)
plt.hist(df['promotion'], bins=20, color='purple', alpha=0.7)
plt.title('Promotion Distribution')

plt.tight_layout()
plt.show()

# Box plots to identify outliers
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.boxplot(y=df['tv_spend'], color='blue')
plt.title('TV Spend')

plt.subplot(2, 2, 2)
sns.boxplot(y=df['radio_spend'], color='green')
plt.title('Radio Spend')

plt.subplot(2, 2, 3)
sns.boxplot(y=df['online_spend'], color='red')
plt.title('Online Spend')

plt.subplot(2, 2, 4)
sns.boxplot(y=df['promotion'], color='purple')
plt.title('Promotion')

plt.tight_layout()
plt.show()

# One-hot encode the season variable
df_encoded = pd.get_dummies(df, columns=['season'], drop_first=True)

# Correlation matrix
plt.figure(figsize=(10, 8))
corr_matrix = df_encoded.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Pair plot to visualize relationships between variables
sns.pairplot(df_encoded)
plt.show()
```
